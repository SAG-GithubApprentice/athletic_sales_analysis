```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames
as2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
as2021 = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2021 sales DataFrame
as2021.head()
```

```python
# Check the 2020 sales data types.
as2020.dtypes
```

```python
# Check the 2021 sales data types.
as2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
as2020_2021 = pd.concat([as2020, as2021], ignore_index=True).reset_index(drop=True)
```

```python
# Check if any values are null.
null_values = as2020_2021.isnull().any()

if as2020_2021.isnull().any().any():
    print("There are null values in the DataFrame.")
else:
    print("There are no null values in the DataFrame.")
```

```python
# Check the data type of each column
as2020_2021.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
as2020_2021['invoice_date'] = pd.to_datetime(as2020_2021['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
as2020_2021.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
grouped_as2020_2021 = as2020_2021.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the sum to "Total_Products_Sold".
grouped_as2020_2021.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
grouped_as2020_2021.head(5)
                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_table_as2020_2021 = pd.pivot_table(as2020_2021, values='units_sold', index=['region', 'state', 'city'], aggfunc='sum').reset_index()

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_table_as2020_2021.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
pivot_table_as2020_2021.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_as2020_2021 = as2020_2021.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()

# Rename the "total_sales" column to "Total Sales"
total_sales_as2020_2021.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
total_sales_as2020_2021.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
ptable_total_sales_as2020_2021 = pd.pivot_table(as2020_2021, values='total_sales', index=['region', 'state', 'city'], aggfunc='sum').reset_index()


# Optional: Rename the "total_sales" column to "Total Sales"
ptable_total_sales_as2020_2021.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
ptable_total_sales_as2020_2021.head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_athletic_footwear_sales = as2020_2021[as2020_2021['product'].str.lower() == "women's athletic footwear"]

# Displayed women's footwear sales data
womens_athletic_footwear_sales.head(5)
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
group_womens_athletic_footwear = womens_athletic_footwear_sales.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
group_womens_athletic_footwear.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Show the top 5 results.
group_womens_athletic_footwear.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
ptable_womens_athletic_footwear = pd.pivot_table(womens_athletic_footwear_sales, 
                                                      values='units_sold', 
                                                      index=['retailer', 'region', 'state', 'city'], 
                                                      aggfunc='sum').reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
ptable_womens_athletic_footwear.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Show the top 5 results.
ptable_womens_athletic_footwear.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
ptable_total_sales_by_date = pd.pivot_table(as2020_2021, 
                                                 values='total_sales', 
                                                 index='invoice_date', 
                                                 aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
ptable_total_sales_by_date.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the table.
ptable_total_sales_by_date.head(5)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_total_sales = ptable_total_sales_by_date.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
sort_daily_total_sales = daily_total_sales.sort_values(by='Total Sales', ascending=True)

sort_daily_total_sales
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_total_sales = ptable_total_sales_by_date.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
sort_weekly_total_sales = weekly_total_sales.sort_values(by='Total Sales', ascending=True)

sort_weekly_total_sales
```
